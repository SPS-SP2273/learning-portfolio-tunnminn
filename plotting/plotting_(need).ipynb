## 1 A simple plot

### 1.1 Letâ€™s look at some code.

```python
from matplotlib import pyplot as plt

# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_2, c ='red', ls='dashed',lw = 4 , label='Y values')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper right')
plt.rcdefaults()
plt.show()
```

```python
plt.plot(x, y_1)
```

### 1.2 Adding another plot

```python
plt.plot(x, y_1, c ='red', ls='dashed',lw = 1 , label='Y values')
plt.plot(x, y_2, c='blue', label = 'Y2 values')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower left')
plt.show()
```

### 1.3 Yet another plot but with error bars

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

plt.plot(x, y_1, c ='red', ls='dashed',lw = 1 , label='Y values')
plt.plot(x, y_2, c='blue', label = 'Y2 values')
plt.errorbar(x, y_3, xerr= x_error, yerr= y_error,
         c = 'green', label = 'Y3 values with error' )
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y', c = 'blue')
plt.grid(alpha=.4)
plt.legend(loc='upper left')
plt.show()
```

### 1.4 Experiment, Experiment, Experiment

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

plt.plot(x, y_1, c ='black', ls='dashed',lw = 1 , label='Y values')
plt.plot(x, y_2, c='pink', label = 'Y2 values')
plt.errorbar(x, y_3, xerr= x_error, yerr= y_error,
         c = 'purple', label = 'Y3 values with error' )
plt.xlabel('x-valu8es')
plt.ylabel('y-val9es')
plt.title('X vs Y', c = 'blue')
plt.grid(alpha=1)
plt.legend(loc = 'lower left')
plt.show()

#acceptable regions for legend are upper and lower left and right
#not specifying a region leads causes the legend to be situated at the generic position of upper left
#the alpha appears to affect the opacity of the grid lines
```

### 1.5 Better with NumPy

```python
import numpy as np
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')
plt.title('Y against X', c = 'red')
plt.grid(alpha = 0.4)
plt.legend()
plt.show()
```

### 1.6 Adding mathematical functions

```python

x = np.array([0, 1, 2, 3, 4, 5, 6])
x2= x**2
sin_x= np.sin(x)

plt.plot(x, x2, c='red', linestyle = 'dashed', label = 'x^2')
plt.plot(x, sin_x, c ='blue', label = 'sin(x)')
plt.title('Sin(x) and X^2')
plt.legend() 
plt.show()
```

### 1.7 We need another axis!

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x = np.linspace(0,19,570)
#linspace syntax --> np.linspace(start, stop, number)
#number refers to the number of samples generated which increases the smoothness of the curve due to morepoints being generated

x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.legend(loc='lower left')

# This creates a new y-axis for the plot that comes after
plt.twinx()
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend(loc='lower right')

plt.show()
```

### 1.8 Saving to disc

```python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.legend(loc='lower left')

# This creates a new y-axis for the plot that comes after
plt.twinx()
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend(loc='lower right')

plt.savefig('simple-plot.png', dpi=150)
plt.savefig('/Users/tunmin/Desktop/SPS/SP2273/simple-plot.png', dpi=160)
```

## 2 A real example: Global Warming

### 2.1 Plotting data from files

```python
data = np.loadtxt('Complete_TAVG_daily.txt', skiprows = 24)
date = data[:,0]
anomaly = data[:,-1]

plt.plot(date, anomaly, alpha=1)
plt.ylim([-8,8])
```

```python
plt.style.use('default')
plt.ylim([-8,8])
plt.plot(date, anomaly, alpha=.5)
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.xlabel('Date')
plt.ylabel('Temperature anomaly')
plt.title('Temperature anomaly\n (Relative to average from Jan 1951 - Dec1980)')
```

### 2.2 xkcd!

```python
plt.xkcd()

plt.ylim([-8,8])
plt.plot(date, anomaly, alpha=.5)
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.xlabel('Date')
plt.ylabel('Temperature anomaly')
plt.title('Temperature anomaly\n (Relative to average from Jan 1951 - Dec1980)')
```

## Exercise 2 :  A simple mathematical plot

```python
x= list(range(0, 51))
print(x)
x = np.array(x)
x = np.linspace(0,50,300)
y = (np.exp(-x/10))*np.sin(x)

plt.xlabel('X')
plt.ylabel('y',rotation = 0)
plt.grid(alpha=.4)
plt.title('y=$e^{-x/10}$sin(x)')
plt.plot(x,y, c = 'blue')
```

```python
x= list(range(0, 51))
print(x)
x = np.array(x)
x = np.linspace(0,50,300)
plt.ylim([-1,1])
y = (np.exp(-x/10))*np.sin(x)

plt.xlabel('X', size = 15, c='blue')
plt.ylabel('y', size= 15, c ='blue', rotation = 0)
plt.grid(alpha=.4)
plt.title('$y=e^{-x/10}\sin(x)$', size =20)
plt.plot(x, y, c = 'blue')
```
