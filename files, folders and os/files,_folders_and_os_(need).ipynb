## 1 Important concepts

### 1.1 Path

The path refers to an address within the computer where the desired file is stored in. 

Absolute path: specified location
Relative path: respective to the location, identify where the desired file is and the relative location to this folder.

### 1.2 More about relative paths

.\data-files\data-01.txt means the file data-01.txt in the folder data-files in the current folder.
..\data-files\data-01.txt means the file data-01.txt in the folder data-files located in the folder above.

. indicates this current folder
.. indicates the folder above

for mac: ~ is used to refer to the home directory.

### 1.3 Path separator

windows uses \
mac and linux uses /


### Text files vs Binary files

.txt files: text files
.md fies: markdown files
.csv files: comma separated values
.xlsx: excel files




### Extentions

file naming convention: name.extension
.extension refers to the file types


### 1.4 Opening and closing files

```python
with open('spectrum-01.txt', 'r') as file:
    file_content = file.read()
#open() opens a stream into the file, r specifies the function to only read the file
#with --> removes the need to manually close the stream once done.
#"r" - Read - Default value. Opens a file for reading, error if the file does not exist

#"a" - Append - Opens a file for appending, creates the file if it does not exist

#"w" - Write - Opens a file for writing, creates the file if it does not exist

#"x" - Create - Creates the specified file, returns an error if the file exist
print(file_content)
```

```python
text = 'Far out in the uncharted backwaters of the unfashionable end of the western spiral arm of the Galaxy lies a small unregarded yellow sun.\nOrbiting this at a distance of roughly ninety-two million miles is an utterly insignificant little blue green planet whose ape-descended life forms are so amazingly primitive that they still think digital watches are a pretty neat idea.'

with open('my-text-once.txt', 'w') as file:
    file.write(text)
    
with open('my-text-once.txt', 'r') as file:
    file_content2 = file.read()
print(file_content2)
```

```python
with open('my-text-lines.txt', 'w') as file:
    for line in text.splitlines():
        file.writelines(line)
```

## 2 Playing with the OS

### 2.1 Some packages

os -> talks to the Operating System to create, modify, delete folders and write OS-agnostic codes

glob -> searching for the files

shutil -> to copy files

shutil.copy(source, destination, *, follow_symlinks = True)
shutil ensures the file permission mode is copied but not its other meta data such as time of creation

shutil allows for higher level functions compared to os copying.

```python
import os
import glob
import shutil
```

### 2.2 OS safe paths

```python
path = os.path.join('.', 'all-data', 'sg-data', 'data-01.txt')
print(path)

#using OS.path.join, it allows the code to run on all OS's by tailoring the path to the specific OS.
# there is a need to know the exact location of the desired path before executing the function.
```

### 2.3 Creating folders

```python
os.mkdir('people')

for person in ['John', 'Paul', 'Ringo']:
    path = os.path.join('people', person)
    print(f'Creating {path}')
    os.mkdir(path)
```

### 2.4 Checking for existence

```python
for person in ['John', 'Paul', 'Ringo']:
    path = os.path.join('people', person)
    try:
        os.mkdir(path)
        print(f"Creating {path}")
    except FileExistsError:
        print(f"{path} already exists; skipping creation")
```

```python
for people in ['John', 'Paul', 'Ringo']:
    path = os.path.join('people', person)
    if os.path.exists(path):
        print(f'{path} already exists, skipping creation')
    else:
        os.mkdir(path)
        print(f'Creating {path}')
```

### 2.5 Copying files

```python

for person in ['John', 'Paul', 'Ringo']:
    path_to_destination = os.path.join('people', person)
    shutil.copy('sp2273_logo.png', path_to_destination)
    print(f"Copied file to {path_to_destination}")
#interating through the for loop, for each individual person, it creates a separate file path
#shutil.copy allows the system to copy this specific file 'sp2273_logo.png' into the created file path for each of the people in persons
```

```python
for person in ['John', 'Paul', 'Ringo']:
    # Create folder 'imgs'
    path_to_imgs = os.path.join('people', person, 'imgs')
    if not os.path.exists(path_to_imgs):
        os.mkdir(path_to_imgs)
#move files to logo
    current_path_for_logo = os.path.join('people', person, 'sp2273_logo.png')
    new_path_for_logo = os.path.join('people', person, 'imgs', 'sp2273_logo.png')
    
    shutil.move(current_path_for_logo, new_path_for_logo)
    print(f'Moved logo to {new_path_for_logo}')
```
