## 3 Housekeeping

### 3.1 Dropping unnecessary columns

```python
df_class2 = df_class.drop(columns=['Unnamed: 0'])
df_class2.head()
#.head() gives the header and first 5 values
```

```python
#using inplace=True
#df_class.drop(columns = ['Unnamed: 0'], inplace = True)
df_class
```

### 3.2 Reorganising columns

```python
columns_to_keep=['Student No', 'Name', 'Major', 'Gender', 'Test 1 (30%)', 'Test 2 (20%)']
#rearranging the columns based on the column header
df_class = df_class[columns_to_keep]
df_class
```

### 3.3 Renaming columns

```python
new_column_info ={'Student No':'Matric_No',
                 'Test 1 (30%)': 'Test 1',
                 'Test 2 (20%)': 'Test 2'}
df_class.rename(columns=new_column_info, inplace=True)
df_class.head()
```

### 3.4 Setting the index

```python
df_class.index

df_class.set_index('Matric_No', drop=False, inplace=True)
df_class.head()
```

### 3.5 Applying changes

```python
print(df_class["Major"].unique())
print(df_class['Gender'].unique())

replace_info = {'PHY':'Physics',
               'CHM':'Chemistry',
               'LS':'Life Sciences',
               'CBIO': 'Computational Biology',
               'M':'Male',
               'F':'Female',
               'NB':'Non-binary'
               }

df_class.replace(to_replace = replace_info, inplace=True)
df_class.head()
```

```python
def clean(text):
    try:
        return text.replace("'", "")
    except AttributeError:
        return text

df_class['Test 2'] = df_class['Test 2'].apply(clean)
df_class.head()
```

```python
df_class.info()
```

### 3.6 Changing Column Type

```python
df_class.info()
new_type_info = {'Major':'category',
                'Gender' : 'category',
                'Test 2': 'float'}

df_class = df_class.astype(new_type_info)

df_class.head()
```

### 3.7 Adding a new column

```python
df_class['Total']= df_class['Test 1'] + df_class['Test 2']
df_class.head()
```

### 3.8 Saving to file

```python
df_class.to_excel('finalised_scores.xlsx', index=True)
```
