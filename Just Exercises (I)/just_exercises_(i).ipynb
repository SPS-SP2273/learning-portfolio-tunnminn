## Exercise 4 :  A function for Fibonacci

```python
def fibonacci(n):
    fibonacci = [0,1]
    for i in range(1, n):
        fibonacci.append(fibonacci[i-1] + fibonacci[i])
    return fibonacci

fibonacci(10)

```

## Exercise 5 :  max_info() without NumPy

```python
numbers = [40, 27, 83, 44, 74, 51, 76, 77, 10, 49]
def max_info(numbers):
    my_max = [70]
    for n in numbers:
        for ele in my_max:
            if n > ele in my_max:
                my_max.append(n)
            else:
                return my_max

numbers = [40, 27, 83, 44, 74, 51, 76, 77, 10, 49]

print(max_info(numbers))
```

```python
numbers = [40, 27, 83, 44, 74, 51, 76, 77, 10, 49]
def max_info(numbers):
    my_max = [30]
    for n in numbers:
        if n > my_max[0]:
            my_max.remove(my_max[0])
            my_max.append(n)
        else:
            return my_max
        
print(max_info(numbers))
```
