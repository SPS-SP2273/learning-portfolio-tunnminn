## Exercise 4 :  A function for Fibonacci

```python
def fibonacci(n):
    fibonacci = [0,1]
    for i in range(1, n):
        fibonacci.append(sum(fibonacci[-2:]))
    assert n>0, 'Fibonacci does not take negative numbers!'
    return fibonacci

fibonacci(9)

```

## Exercise 5 :  max_info() without NumPy

```python
numbers = [40, 27, 83, 44, 74, 51, 76, 77, 10, 49]
def max_info(numbers):
    my_max = [70]
    for n in numbers:
        for ele in my_max:
            if n > ele in my_max:
                my_max.append(n)
            else:
                return my_max

numbers = [40, 27, 83, 44, 74, 51, 76, 77, 10, 49]

print(max_info(numbers))
```

```python
numbers = [40, 27, 83, 44, 74, 51, 76, 77, 10, 49]
def max_info(numbers):
    my_max = [30]
    for n in numbers:
        if n > my_max[0]:
            my_max.remove(my_max[0])
            my_max.append(n)
        else:
            return my_max
        
print(max_info(numbers))
```

```python
numbers = [40, 27, 83, 44, 74, 51, 76, 77, 10, 49]

def max_info(numbers):
    my_max = numbers[0]
#second variable for the index of my_max
    my_max_index = 0
    for number in numbers:
        if my_max < number:
            my_max = number
#outside of the if loop, to append the index of the max number
        my_max_index = numbers.index(my_max)
    return my_max, my_max_index

max_info(numbers)
```
